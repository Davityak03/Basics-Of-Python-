q1: Read Total profit of all months and show it using a line plot

    X label name = Month Number
    
    Y label name = Total profit

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('company_sales_data.csv')
arr1=np.array(df['total_profit'])
arr2=np.array(df['month_number'])
month_number=df['month_number'].tolist()
plt.plot(arr2,arr1,'b-s')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.xticks(month_number)
plt.show()
```

q3: Read all product sales data and show it  using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('company_sales_data.csv')
month_number=np.array(df['month_number'])
facecream=np.array(df['facecream'])
facewash=np.array(df['facewash'])
toothpaste=np.array(df['toothpaste'])
bathingsoap=np.array(df['bathingsoap'])
shampoo=np.array(df['shampoo'])
moisturizer=np.array(df['moisturizer'])
plt.plot(month_number,facecream,label='face cream data',marker='o')
plt.plot(month_number,facewash,label='facewash data',marker='o')
plt.plot(month_number,toothpaste,label='toothpaste data',marker='o')
plt.plot(month_number,bathingsoap,label='bathingsoap data',marker='o')
plt.plot(month_number,shampoo,label='shampoo data',marker='o')
plt.plot(month_number,moisturizer,label='moisturizer data',marker='o')
plt.legend(loc='upper left')
plt.xlabel('Month Number')
plt.ylabel('Sales Data')
plt.xticks(month_number)
plt.show()
```

q4: Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

```python
arr1=np.array(df['toothpaste'])
arr2=np.array(df['month_number'])
plt.scatter(arr2,arr1)
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales Data')
plt.grid(True,linestyle="--")
plt.xticks(month_number)
plt.show()
```

q5: Read face cream and facewash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
arr1=np.array(df['facecream'])
arr2=np.array(df['facewash'])
arr3=np.array(df['month_number'])
plt.bar([a-0.25 for a in arr3], arr1, width= 0.25, label='Face Cream sales data',align='edge')
plt.bar([a+0.25 for a in arr3], arr2, width=-0.25,label='Face Wash sales data',align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units')
plt.legend(loc='upper right')
plt.xticks(month_number)
plt.show()
```

q7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
profitList=np.array(df ['total_profit'])
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(profitList, profit_range, label = 'Profit data',rwidth=0.7)
plt.xlabel('profit range')
plt.ylabel('Actual Profit')
plt.legend(loc='upper right')
plt.xticks(profit_range)
plt.title('Profit data')
plt.show()
```

q8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
labels=['Facecream','Facewash','Toothpaste','Bathingsoap','Shampoo','Moisturizer']
data_of_sales=[df["facecream"].sum(),df["facewash"].sum(),df["toothpaste"].sum(),df["bathingsoap"].sum(),df["shampoo"].sum(),df["moisturizer"].sum()]
plt.pie(data_of_sales,labels=labels,autopct='%1.1f%%')
plt.axis("equal")
plt.legend(loc='lower right')
plt.show()
```

q10: Read all product sales data and show it using the stack plot

```python
month_number=np.array(df['month_number'])
facecream=np.array(df['facecream'])
facewash=np.array(df['facewash'])
toothpaste=np.array(df['toothpaste'])
bathingsoap=np.array(df['bathingsoap'])
shampoo=np.array(df['shampoo'])
moisturizer=np.array(df['moisturizer'])
plt.plot([],[],color='b', label='face Cream', linewidth=5)
plt.plot([],[],color='r', label='Face wash', linewidth=5)
plt.plot([],[],color='g', label='Tooth paste', linewidth=5)
plt.plot([],[],color='k', label='Bathing soap', linewidth=5)
plt.plot([],[],color='c', label='Shampoo', linewidth=5)
plt.plot([],[],color='y', label='Moisturizer', linewidth=5)
plt.stackplot(month_number,facecream,facewash,toothpaste,bathingsoap,shampoo,moisturizer,colors=['m','c','r','k','g','y'])
plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.xticks(month_number)
plt.show()
```
